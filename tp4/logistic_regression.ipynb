# Régression logistique (classification des fleurs d’iris)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
#chargement de base de données iris 
iris = datasets.load_iris() 
```

```python
# choix de deux variables : Utiliser les deux premières colonnes afin d'avoir un problème de classification binaire
X = iris.data[:, :2]
y = (iris.target != 0) * 1 # re-étiquetage des fleurs
```

```python
# Visualisation
plt.figure(figsize=(7,5))
plt.scatter(X[y==0, 0], X[y==0, 1], color="blue", label="Iris Setosa (classe 0)")
plt.scatter(X[y==1, 0], X[y==1, 1], color="red", label="Non-Setosa (classe 1)")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Visualisation des données Iris (2 features)")
plt.legend()
plt.show()
```

✔ On remarque que Setosa (classe 0) est très bien séparée du reste.
Elle possède :

des sépales plus courts,

et plus larges → ce qui la rend très distincte.

✔ Les fleurs non-Setosa (classe 1), c’est-à-dire Versicolor & Virginica,
sont mélangées, car elles ont des mesures proches.

Conclusion :

→ Le problème binaire Setosa vs autres est facilement linéairement séparable.

→ Une régression logistique est donc très adaptée.

```python
model = LogisticRegression(C=1e20)
model.fit(X, y)
```

```python
test_points = np.array([[5.5, 2.5],
                        [7.0, 3.0],
                        [3.0, 2.0],
                        [5.0, 3.0]])

pred = model.predict(test_points)
print(pred)
```
